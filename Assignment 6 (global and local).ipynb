```python
#1
def show_number():
    x = 10      # local variable
    print("Inside function:", x)

show_number()

```

```python
#2
def payment_details():
    global total
    total = 2299.0
    print("Total payment:", total)

payment_details()
print("Access outside:", total)
```

```python
#3
count = 0

def increment():
    global count
    count += 1

increment()
increment()
increment()
print("Final count:", count)

```

```python
#4
total_sales = 0

def sale_day1(amount):
    global total_sales
    total_sales += amount

def sale_day2(amount):
    global total_sales
    total_sales += amount

sale_day1(1000)
sale_day2(1500)
print("Total sales:", total_sales)
```

```python
#6
def outer():
    x = 5

    def inner():
        nonlocal x
        x += 10

    inner()
    print(x)

outer()
```

```python
#7
g = "Global"

def outer():
    e = "Enclosing"

    def middle():
        l = "Local"
        print(g)
        print(e)
        print(l)

    middle()

outer()
```

```python
#12
def marks(m1, m2, m3):
    def total():
        return m1 + m2 + m3
    print("Total marks:", total())

marks(70, 80, 90)
```

```python
#13
def counter():
    count = 0

    def increase():
        nonlocal count
        count += 1
        print(count)

    increase()
    increase()
    increase()

counter()
```

```python
#14
def multiplier(n):
    def multiply(x):
        return x * n
    return multiply

double = multiplier(2)
print(double(5))
print(double(10))
```

```python
#15
def add(*numbers):
    print(sum(numbers))

add(1, 2, 3, 4)
```

```python
#23
def employee(name, salary):
    def bonus():
        return salary * 0.10
    print(name, "Bonus:", bonus())

employee("Amit", 50000)
```

```python
#24
def calls():
    count = 0

    def track():
        nonlocal count
        count += 1
        print("Called", count, "times")

    track()
    track()
    track()

calls()
```

```python
#25
data = "Global data"

def show():
    data = "Local data"
    print(data)

show()
print(data)

```

```python
#26
def expenses():
    total = 0

    def add(amount):
        nonlocal total
        total += amount
        print("Total:", total)

    add(100)
    add(200)

expenses()
```

```python
#27
def join_strings(*words):
    print(" ".join(words))

join_strings("Hello", "Python", "World")
```

```python
#28
def config(**settings):
    for k, v in settings.items():
        print(k, ":", v)

config(OS="Windows", RAM="8GB")

```

```python
#29
def order(*items, **prices):
    print("Items:", items)
    print("Prices:", prices)

order("Pen", "Book", Pen=10, Book=50)
```

```python
#30
def salary(base):
    total = base

    def incentive(amount):
        nonlocal total
        total += amount
        print("Total salary:", total)

    incentive(2000)
    incentive(3000)

salary(20000)
```
