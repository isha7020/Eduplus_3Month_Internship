```python
#2
dict_student={
    'student_Name':['mukta','aishwarya','sanika','mahek','bhakti'],
    'Math':[ 60,65,70,75,80],
    'Science':[70,65,80,75,60],
    'English':[60,56,78,80,70]
}

df=pd.DataFrame(dict_student)
print(df)
top_3=df.head(3)
print("\ntop 3 rows in dataframe \n: ", top_3)
df.columns = df.columns.str.strip()
print(df.columns)
    
    
    
```

```python
#4
df['Total_marks']=df['Math']+df['Science']+df['English']
df['Average']=df['Total_marks']/3
print(df)
```

```python
#5
df.loc[1,'Science']=None
print("Missing value:",df)
print(df.isnull())
df['Science'].fillna(df['Science'].mean(),inplace=True)

```

```python
#6
print(df[df['Average']>75])
```

```python
#7
df_sorted=df.sort_values(by="Total_marks",ascending=False)
top_3 = df_sorted.head(3)
print(top_3)
```

```python
df_marks = pd.DataFrame({
    'Student_Name': ['Amit', 'Neha', 'Rahul', 'Sneha'],
    'Total_Marks': [250, 270, 260, 280]
})


df_sports = pd.DataFrame({
    'Student_Name': ['Amit', 'Neha', 'Rahul', 'Sneha'],
    'Sports_Score': [20, 25, 18, 30]
})
merged_df = pd.merge(df_marks, df_sports, on='Student_Name')
print(merged_df)
merged_df['Overall_Score'] = merged_df['Total_Marks'] + merged_df['Sports_Score']
print(merged_df)

```

```python
# Define Desktop path
import os
desktop_path = os.path.join(os.path.expanduser("~"), "Desktop")

# File path
file_path = os.path.join(desktop_path, "final_student_scores.csv")

# Export to CSV
merged_df.to_csv(file_path, index=False)

if os.path.exists(file_path):
    print("File successfully created on Desktop.")
else:
    print("File not found.")

```
