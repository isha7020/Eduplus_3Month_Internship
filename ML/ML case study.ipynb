```python
import pandas as pd

products = ["Mobile", "Laptop", "Mouse", "Keyboard", "Headphones"]
reviews=["good","average","excellent"]
df = pd.DataFrame({
    "Order_Id": [x for x in range(1, 51)],
    "Customer_Id": [100 + x for x in range(50)],
    "Product": [products[i % 5] for i in range(50)],
    "Category": ["Electronics"]*50,
    "Quantity": [(i % 3) + 1 for i in range(50)],
    "Price": [15000, 50000, 500, 1200, 2500] * 10,
    "Rating": [(i % 5) + 1 for i in range(50)],
    "OrderDate": pd.date_range(start="2023-01-01", periods=50, freq="7D"),
    "Review":[reviews[i%3] for i in range(1,51)]

})

print(df)
```

```python
#Q8: Find the top 5 best-selling products based on total quantity sold
# Display the product name and total quantity in descending order
product_sales = df.groupby("Product")["Quantity"].sum()
product_sales = product_sales.sort_values(ascending=False)
top_5_products = product_sales.head(5)
top_5_products
```

```python
#9
category_summary = df.groupby("Category").agg(
    Total_Sales=("Total_Amount", "sum"),
    Average_Rating=("Rating", "mean"),
    Number_of_Orders=("Category", "count")
)

category_summary = category_summary.sort_values(
    by="Total_Sales",
    ascending=False
)

category_summary
```

```python
# Q10: Create a pivot table showing average Price for each Category and Rating combination
# Use groupby with multiple columns (Category, Product) and aggregate multiple functions
# Pivot Table
pivot_table = pd.pivot_table(
    df,
    values="Price",
    index="Category",
    columns="Rating",
    aggfunc="mean"
)


grouped_data = df.groupby(["Category", "Product"]).agg(
    Average_Price=("Price", "mean"),
    Total_Quantity=("Quantity", "sum"),
    Order_Count=("Product", "count")
)

pivot_table
grouped_data

```
