```python
#4
v = tf.Variable(5.0)
print("Initial Value:",v.numpy())
```

```python
#6
matrix_var = tf.Variable([[1.0, 2.0],
                           [3.0, 4.0]])
matrix_var.assign_add([[1.0, 1.0], [1.0, 1.0]])
print("After assign_add():\n", matrix_var.numpy())
```

```python
#7

x = tf.Variable(3.0)

with tf.GradientTape() as tape:
    y = x**2 + 2*x + 1

print("x value:", x.numpy())
print("y value:", y.numpy())

```

```python
#8
dy_dx = tape.gradient(y, x)
print("Gradient dy/dx:", dy_dx.numpy())

```

```python
#9
a = tf.constant([[1, 2],
                 [3, 4]])
b = tf.constant([[5, 6],
                 [7, 8]])

matmul_result = tf.matmul(a, b)
print("Matrix multiplication result:\n", matmul_result.numpy())
```
