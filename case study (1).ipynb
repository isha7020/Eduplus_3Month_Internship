```python
#B
w1={
    "milk":29, "biscuit":10}
w2={"bread":50,"milk":20}


def common_product(w1,w2):
    return set(w1.keys()) & set( w2.keys())
def unique_product(w1,w2):
    return set(w1.keys()) | set(w2.keys())
def combined_product(w1,w2):
    combined={}
    combined.update(w1)
    combined.update(w2)
    return sorted(combined.items(),key= lambda x:x[1],reverse=True)

print("All COMMON PRODUCTS:",common_product(w1,w2))
print("ALL UNIQUE PRODUCTS:",unique_product(w1,w2))
print("descending order of quantity:",combined_product(w1,w2))
```

```python
def non_repeated_courses(student):
    all_courses = []

    for courses in student.values():
        for c in courses:
            all_courses.append(c)

    print("\nCourses not repeated by any student:")
    for c in set(all_courses):
        if all_courses.count(c) == 1:
            print(c)
```

```python
#F
# Take attendance for 5 days
def get_attendance():
    day1 = set(input("Day 1 students: ").split(","))
    day2 = set(input("Day 2 students: ").split(","))
    day3 = set(input("Day 3 students: ").split(","))
    day4 = set(input("Day 4 students: ").split(","))
    day5 = set(input("Day 5 students: ").split(","))

    return day1, day2, day3, day4, day5


# Students present every day
def present_every_day(d1, d2, d3, d4, d5):
    print("Present every day:", d1 & d2 & d3 & d4 & d5)


# Students absent at least one day
def absent_at_least_one_day(d1, d2, d3, d4, d5):
    all_students = d1 | d2 | d3 | d4 | d5
    present_all = d1 & d2 & d3 & d4 & d5
    print("Absent at least one day:", all_students - present_all)


# Total unique students
def total_unique_students(d1, d2, d3, d4, d5):
    all_students = d1 | d2 | d3 | d4 | d5
    print("Total unique students:", len(all_students))


# Function calls
d1, d2, d3, d4, d5 = get_attendance()
present_every_day(d1, d2, d3, d4, d5)
absent_at_least_one_day(d1, d2, d3, d4, d5)
total_unique_students(d1, d2, d3, d4, d5)
```

```python
#5

def get_purchases():
    purchases = []
    n = int(input("Enter number of customers: "))

    for i in range(n):
        name = input("Enter customer name: ")
        products = input("Enter products (comma separated): ").split(",")
        products = [p.strip() for p in products]
        purchases.append((name, products))

    return purchases


# 1. Total number of unique products
def total_unique_products(purchases):
    unique_products = set()

    for customer, products in purchases:
        for p in products:
            unique_products.add(p)

    print("\nTotal unique products bought:", len(unique_products))


# 2. Customers who bought both Laptop and Headphones
def laptop_and_headphones(purchases):
    print("\nCustomers who bought both Laptop and Headphones:")
    for customer, products in purchases:
        if "Laptop" in products and "Headphones" in products:
            print(customer)


# 3. Most frequently purchased product(s)
def most_frequent_products(purchases):
    product_count = {}

    for customer, products in purchases:
        for p in products:
            if p in product_count:
                product_count[p] += 1
            else:
                product_count[p] = 1

    max_count = max(product_count.values())

    print("\nMost frequently purchased product(s):")
    for product, count in product_count.items():
        if count == max_count:
            print(product)


# Function calls
purchases = get_purchases()
total_unique_products(purchases)
laptop_and_headphones(purchases)
most_frequent_products(purchases)
```

```python
#6


# 1. Function to take booking details from user
def get_bookings():
    bookings = []

    n = int(input("Enter number of bookings: "))

    for i in range(n):
        print(f"\nEnter details for booking {i+1}")
        name = input("Traveler Name: ")
        destination = input("Destination: ")
        date = input("Date (YYYY-MM-DD): ")

        booking = (name, destination, date)
        bookings.append(booking)

    return bookings


# 2. Function to remove duplicate records using set
def remove_duplicates(bookings):
    unique_bookings = set(bookings)
    return unique_bookings


# 3. Function to display all unique destinations
def show_unique_destinations(bookings):
    destinations = set()

    for booking in bookings:
        destinations.add(booking[1])  # destination is at index 1

    print("\nUnique Destinations:")
    for dest in destinations:
        print(dest)


# 4. Function to find top 3 most visited destinations
def top_3_destinations(bookings):
    count = {}

    for booking in bookings:
        destination = booking[1]
        count[destination] = count.get(destination, 0) + 1

    # Sort destinations by frequency (highest first)
    sorted_dest = sorted(count.items(), key=lambda x: x[1], reverse=True)

    print("\nTop 3 Most Visited Destinations:")
    for dest, freq in sorted_dest[:3]:
        print(dest, "→", freq, "visits")



bookings = get_bookings()

unique_bookings = remove_duplicates(bookings)

print("\nBookings after removing duplicates:")
for b in unique_bookings:
    print(b)

show_unique_destinations(unique_bookings)

top_3_destinations(unique_bookings)
```

```python
#7
# 1. Function to take user movie data
def get_user_movies():
    users = {}

    n = int(input("Enter number of users: "))

    for i in range(n):
        print(f"\nEnter details for user {i+1}")
        name = input("User Name: ")

        movies = input("Enter favorite movies (comma separated): ")
        movie_tuple = tuple(m.strip() for m in movies.split(","))

        users[name] = movie_tuple

    return users


# 2. Function to find users sharing at least two movies
def find_common_users(users):
    print("\nUsers sharing at least 2 common movies:")

    user_list = list(users.keys())

    for i in range(len(user_list)):
        for j in range(i + 1, len(user_list)):
            user1 = user_list[i]
            user2 = user_list[j]

            common_movies = set(users[user1]) & set(users[user2])

            if len(common_movies) >= 2:
                print(user1, "and", user2, "→ Common Movies:", common_movies)


# 3. Function to display all unique movies
def show_unique_movies(users):
    all_movies = set()

    for movie_list in users.values():
        all_movies.update(movie_list)

    print("\nAll Unique Movies:")
    for movie in all_movies:
        print(movie)


# 4. Function to find movies liked by only one user
def movies_liked_by_one_user(users):
    movie_count = {}

    for movie_list in users.values():
        for movie in movie_list:
            movie_count[movie] = movie_count.get(movie, 0) + 1

    print("\nMovies liked by only one user:")
    for movie, count in movie_count.items():
        if count == 1:
            print(movie)


# ================= MAIN PROGRAM =================

users = get_user_movies()

find_common_users(users)

show_unique_movies(users)

movies_liked_by_one_user(users)
```

```python
# 1. Function to take participant and sports input
def get_participants():
    participants = {}

    n = int(input("Enter number of participants: "))

    for i in range(n):
        print(f"\nEnter details for participant {i+1}")
        name = input("Participant Name: ")

        sports = input("Enter sports (comma separated): ")
        sport_list = [s.strip() for s in sports.split(",")]

        participants[name] = sport_list

    return participants


# 2. Function to display participants playing both Cricket and Football
def cricket_and_football_players(participants):
    print("\nParticipants playing both Cricket and Football:")

    for name, sports in participants.items():
        sport_set = set(sports)
        if "Cricket" in sport_set and "Football" in sport_set:
            print(name)


# 3. Function to find total number of unique sports
def total_unique_sports(participants):
    all_sports = set()

    for sports in participants.values():
        all_sports.update(sports)

    print("\nTotal unique sports being played:", len(all_sports))
    print("Unique Sports:", all_sports)


# 4. Function to create tuple of sports with highest participation
def highest_participation_sports(participants):
    sport_count = {}

    for sports in participants.values():
        for sport in sports:
            sport_count[sport] = sport_count.get(sport, 0) + 1

    max_count = max(sport_count.values())

    top_sports = tuple(
        sport for sport, count in sport_count.items() if count == max_count
    )

    print("\nSports with highest participation:", top_sports)


# ================= MAIN PROGRAM =================

participants = get_participants()

cricket_and_football_players(participants)

total_unique_sports(participants)

highest_participation_sports(participants)
```
